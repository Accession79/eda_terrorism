### Selecting necessary columns only

### 8. Defining a function for country-wise analysis

```python
# Country-wise analysis of Terror Attacks

def country_wise_analysis(data , Country=None, line_color=None, palette=None, title =None):
    plt.figure(figsize = (16, 12))
    
    
    #Plot 1
    plt.subplot(2,2,1)
    sns.countplot(y = data.loc[data['Country'] == Country]['State'] ,
                  order = data.loc[data.Country == Country].State.value_counts()[:15].index,
                  data = data,
                  palette = sns.color_palette(palette,10),
                  edgecolor = sns.color_palette('dark',5))
    plt.xticks(rotation = 90)
    plt.title(f"15 Most attacked States of {Country}",weight = 'bold' ,fontsize = 18)
    plt.xlabel('Count',fontsize = 20)
    plt.ylabel('State',fontsize = 20)
    
    
    #Plot 2
    plt.subplot(2,2,2)
    sns.countplot(y = data.loc[data['Country'] == Country]['City'] ,
                  order = data.loc[data.Country == Country].City.value_counts()[:15].index,
                  data = data,
                  palette = sns.color_palette(palette,10),
                  edgecolor = sns.color_palette('dark',5))
    plt.xticks(rotation = 90)
    plt.title(f"15 Most attacked cities of {Country}",weight = 'bold' ,fontsize = 18)
    plt.xlabel('Count',fontsize = 20)
    plt.ylabel('City',fontsize = 20)
    

    
    df['Casualities'] = df['Killed'] + df['Attacked']
    
    #Plot 3
    plt.subplot(2,2,3)
    data.query('Country == @Country').groupby(['Year'])['Casualities'].sum().plot(xticks = data.loc[data.Country == Country].groupby(['Year'])['Casualities'].sum().index, rot=90 , color = line_color)
    plt.title(f"Total casualities from (1970 - 2017) in {Country}", fontdict = {'fontsize' : 18 , 'weight': 'bold' } , loc = 'left')
    plt.ylabel("Casualities" ,fontsize = 20)
    plt.xlabel('Year',fontsize = 20)
    
    
    #Plot 4
    plt.subplot(2,2,4)
    sns.countplot(x = data.query('Country == @Country').Group,
                  data = data,
                  order = data.query('Country == @Country').Group.value_counts()[:15].index[1:],
                  palette = sns.color_palette(palette,7),
                  edgecolor = sns.color_palette('dark',5))
    plt.title(f"Terrorist Groups active/were active in {Country}",weight = 'bold' ,fontsize = 18 , loc = 'left')
    plt.xticks(rotation =90)
    plt.xlabel('Group',fontsize = 20)
    plt.ylabel('Attack Count',fontsize = 20)
    
    
    plt.subplots_adjust(wspace = 1 ,hspace =0.5 ,top = 0.5 , bottom = 0.1)  # tune the subplot layout
    plt.tight_layout()   #to avoid overlapping of subplots and fit them cleanly in figure
    plt.show()
```

**- Country-wise analysis for most attacked country 'Iraq'**

```python
country_wise_analysis(df, Country ='Iraq', title = "Success Rate of Terrorist Attacks in Iraq", palette = 'inferno', line_color = 'crimson')
```

**- Country-wise analysis for our country 'India'**

```python
country_wise_analysis(df, Country ='India', title = "Success Rate of Terrorist Attacks in India", palette = 'inferno', line_color = 'crimson')
```

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region each Year', fontsize=20, weight='bold')
plt.ylabel('Number of Attacks', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.show()
```
